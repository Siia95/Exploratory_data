```python
#Прочитайте csv файл (використовуйте функцію read_csv)
import pandas as pd
import matplotlib.pyplot as plt

file_id = '1ERmZE5goTbYXpj5QJUj9omClqW2St3Ml'

url = f'https://drive.google.com/uc?id={file_id}'
df = pd.read_csv(url)
df
```

```python
#Виведіть перші п'ять рядків (використовується функція head)
print("Перші п'ять рядків:")
df.head()
```

```python
#Виведіть розміри датасету (використовуйте атрибут shape)
print("\nРозміри датасету:")
df.shape
```

```python
#Відповідь: Про скільки книг зберігає дані датасет?
print("\nВідповідь: Датасет містить інформацію про", df.shape[0], "книг.")
```

```python
df.columns = ['name', 'author', 'user_rating', 'reviews', 'price', 'year', 'genre']
print("\nНові назви стовпців:")
df.head()
```

```python
#Перевірити пропуски
missing_data = df.isna().sum()
print("\nКількість пропусків у кожному стовпці:")
missing_data
```

```python
any_missing = any(missing_data > 0)
print("\nЧи є в якихось змінних пропуски? Одержана відповідь:")
print("Так" if any_missing else "Ні")
```

```python
#Унікальні значення у стовпці 'genre'
unique_genres = df['genre'].unique()

print("\nУнікальні жанри:")
unique_genres
```

```python
#Побудувати гістограму цін
df['price'].plot(kind='hist', bins=20, title='Розподіл цін')
plt.xlabel('Ціна')
plt.ylabel('Кількість книг')
plt.show()
```

```python
#Визначити максимальну, мінімальну, середню та медіанну ціну
max_price = df['price'].max()
min_price = df['price'].min()
mean_price = df['price'].mean()
median_price = df['price'].median()

print("\nМаксимальна ціна:", max_price)
print("Мінімальна ціна:", min_price)
print("Середня ціна:", mean_price)
print("Медіанна ціна:", median_price)
```

```python
max_rating = df['user_rating'].max()
print("Відповідь: Найвищий рейтинг у датасеті:")
max_rating
```

```python
num_books_max_rating = df[df['user_rating'] == max_rating].shape[0]
print("Відповідь: Кількість книг із найвищим рейтингом:")
num_books_max_rating
```

```python
max_reviews_book = df[df['reviews'] == df['reviews'].max()]['name'].values[0]
print("Відповідь: Книга з найбільшою кількістю відгуків:")
max_reviews_book
```

```python
df_2015 = df[df['year'] == 2015]
df_2015
```

```python
most_expensive_2015 = df_2015[df_2015['price'] == df_2015['price'].max()]['name'].values[0]
print("Відповідь: Найдорожча книга у 2015 році:")
most_expensive_2015
```

```python
num_fiction_2010 = df[(df['genre'] == 'Fiction') & (df['year'] == 2010)].shape[0]
print("Відповідь: Кількість книг жанру Fiction у Топ-50 у 2010 році:")
num_fiction_2010
```

```python
num_high_rating_2010_2011 = df[(df['user_rating'] == 4.9) & ((df['year'] == 2010) | (df['year'] == 2011))].shape[0]
print("Відповідь: Кількість книг із рейтингом 4.9 у 2010 та 2011 роках:")
num_high_rating_2010_2011
```

```python
sorted_books_2015 = df[(df['year'] == 2015) & (df['price'] < 8)].sort_values(by='price')
sorted_books_2015
```

```python
print("\nВідповідь:")
print("Книга остання у відсортованому списку:", sorted_books_2015['name'].iloc[-1])
```

```python
genre_prices = df.groupby('genre')['price'].agg(['max', 'min'])
print("\nВідповідь:")
print("Максимальна ціна для жанру Fiction:", genre_prices.loc['Fiction', 'max'])
print("Мінімальна ціна для жанру Fiction:", genre_prices.loc['Fiction', 'min'])
print("Максимальна ціна для жанру Non Fiction:", genre_prices.loc['Non Fiction', 'max'])
print("Мінімальна ціна для жанру Non Fiction:", genre_prices.loc['Non Fiction', 'min'])
```

```python
author_book_count = df.groupby('author')['name'].count()
print("\nВідповідь:")
print("Розмірність таблиці (кількість рядків):", author_book_count.shape[0])
print("Автор із найбільшою кількістю книг:", author_book_count.idxmax())
print("Кількість книг цього автора:", author_book_count.max())
```

```python
author_avg_rating = df.groupby('author')['user_rating'].mean()
print("\nВідповідь:")
print("Автор із мінімальним середнім рейтингом:", author_avg_rating.idxmin())
print("Середній рейтинг цього автора:", author_avg_rating.min())
```

```python
author_stats = pd.concat([author_book_count, author_avg_rating], axis=1, keys=['book_count', 'avg_rating'])
author_stats
```

```python
sorted_author_stats = author_stats.sort_values(by=['book_count', 'avg_rating'])
sorted_author_stats
```

```python
print("\nВідповідь:")
print("Перший автор у списку:", sorted_author_stats.index[0])
```

```python
#Графік 1: Розподіл рейтингів
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

plt.figure(figsize=(10, 6))
sns.histplot(df['user_rating'], bins=20, kde=True, color='skyblue')
plt.title('Розподіл рейтингів книг')
plt.xlabel('Рейтинг')
plt.ylabel('Частота')
plt.show()
```

```python
#Графік 2: Розподіл цін
plt.figure(figsize=(10, 6))
sns.histplot(df['price'], bins=20, kde=True, color='lightgreen')
plt.title('Розподіл цін на книги')
plt.xlabel('Ціна')
plt.ylabel('Частота')
plt.show()
```

```python
#Графік 3: Кількість книг у кожному жанрі
plt.figure(figsize=(10, 6))
sns.countplot(x='genre', data=df, palette='viridis')
plt.title('Кількість книг у кожному жанрі')
plt.xlabel('Жанр')
plt.ylabel('Кількість книг')
plt.show()
```
